### Build a DNAStringSet out of microarray probes and use the "replaceAt" function to modify it.

```r
library(hgu95av2probe, verbose=FALSE)
probes <- DNAStringSet(hgu95av2probe)
probes
```

We can replace specific bases in the DNA strings with our specific value of interest:

```r
replaceAt(probes, at=IRanges(3, 4), value="-++-") # replace positions 3 and 4
replaceAt(probes, at=IRanges(3, 4), value="") # remove positions 3 and 4
```

The following is a function to plot ranges (adapted from https://web.stanford.edu/class/bios221/)

```r
plotRanges = function(x, xlim=x, main=deparse(substitute(x)), sep=0.5){
  height = 1
  if (is(xlim, "Ranges"))
    xlim = c(min(start(xlim))-sep, max(end(xlim))+sep)
  bins = disjointBins(IRanges(start(x), end(x)+1))  
  
  plot.new()
  plot.window(xlim, c(0, max(bins) * (height + sep)))
  ybottom = bins * (sep + height) - height
  rect(start(x), ybottom, end(x), ybottom + height, col = "blue")
  title(main)
  axis(1)
}
```

```r
par(mfrow=c(2,1))
plotRanges(ir)
plotRanges(ir3)
```

### GRangesList:

Let's start by building our first GRangesList.
To do that, copy the function at: https://github.com/antonio-mora/vennRanges/blob/master/R/generate_random_genomic_range.R

```r
generate_random_genomic_range <- function(number_chrom=5, size_range=c(50, 1000), size_inter_range=c(50, 2000), number_ranges_per_chrom=20) {

	bed_file <- NULL
	for (i in 1:number_chrom) {
		# Generate random numbers:
		range_sizes <- sample(size_range[1]:size_range[2], number_ranges_per_chrom, replace=F)				# all sizes of ranges
		inter_range_sizes <- sample(size_inter_range[1]:size_inter_range[2], number_ranges_per_chrom, replace=F)	# all sizes between ranges

		# Build file range by range:
		old_start <- 0; old_end <- 0; 
		for (j in 1:number_ranges_per_chrom) {
			new_start <- old_end + inter_range_sizes[i]
			new_end <- new_start + range_sizes[i]

			bed_file <- rbind(bed_file, c(paste("chr",i,sep=""), new_start, new_end))

			old_end <- new_end
		}
	}

	# Final touches to resulting data frame:
	bed_file <- as.data.frame(bed_file)
	colnames(bed_file) <- c("chr", "start", "end")
	bed_file$start <- as.numeric(as.character(bed_file$start))
	bed_file$end <- as.numeric(as.character(bed_file$end))

	bed_file
}
```

```r
bed1 = generate_random_genomic_range()
bed2 = generate_random_genomic_range()
bed3 = generate_random_genomic_range()
```

Build a GRangesList with the generated bed files:

```r
gr1 = GRanges(bed1)
gr2 = GRanges(bed2)
gr3 = GRanges(bed3)

grlist = GRangesList(gr1, gr2, gr3)
```

We can use GRanges methods on GRangesLists. For example, let's shift all ranges in grlist by 10:

```r
grlist2 = shift(grlist, 10)
names(grlist2) = c("shiftgr1", "shiftgr2", "shiftgr3")
```

## 2.3) GenomicFeatures:

```r
# GenomicFeatures retrieves and manages different genomic annotations from UCSC and BioMart.
# The TxDB object stores transcript annotations. It does use:
# makeTxDbFromUCSC to download from UCSC Genome browser
# makeTxDbFromBiomart to download from BioMart
# makeTxDbFromGFF (use a GFF3 or GTF file containing transcript information)
```

```r
library(GenomicFeatures)
```

First, let us check what are the supported tables:

```r
supportedUCSCtables()
```

```r
supportedUCSCtables(genome="hg38")
```

Let's build our first Txdb object (this can take a while):

```r
txdb = makeTxDbFromUCSC(genome="hg38",tablename="knownGene")
txdb
```

```r
columns(txdb)
```

transcripts() can be used to explore and filter the transcripts in the database:

```r
transcripts(txdb)
```

```r
transcripts(txdb, filter=list(tx_chrom="chr1"))
```

```r
GR <- transcripts(txdb, filter=list(tx_chrom = "chr15", tx_strand = "+"))
length(GR)
```

transcriptsBy() can be used to select transcripts by gene:

```r
transcriptsBy(txdb, by="gene")
```

exons() can be used for the corresponding exons:

```r
exons(txdb)
```

It is also possible to select the transcripts that overlap with a user-specified region:

```r
gr = GRanges(seqnames= Rle("chr1",  2),ranges=IRanges(start=c(100000,500000), end = c(200000,600000)))
overlap = transcriptsByOverlaps(txdb,  gr)
overlap
```

How many transcripts were found in this genomic region?

promoters() can be also used with additional size parameters:

```r
PR <- promoters(txdb, upstream=2000, downstream=400)
PR
```

Finally, as in other databases, we can use the SELECT command. For example, to extract TXNAME of given GENEIDs:

```r
keys <- c("100033416", "100033417", "100033420")
select(txdb, keys = keys, columns="TXNAME", keytype="GENEID")
```

Or to select multiple values at the same time:

```r
cols <- c("TXNAME", "TXSTRAND", "TXCHROM")
select(txdb, keys=keys, columns=cols, keytype="GENEID")
```
